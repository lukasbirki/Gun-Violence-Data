## 1.2 Exploring Time Related Trends of Gun Violence

```python
df['datetime'] = pd.to_datetime(df['date'])
```

First: Extracting the number of 
- incidents (1 incident = 1 observation)
- people injured 
- people killed 

and storing them in a new dataframe 'df_trend'

```python
df_kills = df['n_killed'].groupby(by = df['datetime']).sum()  #Matrix of Absolute Cases
df_harmed = df['n_injured'].groupby(by = df['datetime']).sum()  #Matrix of Absolute Cases
df_cases = df['incident_id'].groupby(by = df['datetime']).count()  #Matrix of Absolute Cases

df_cases = pd.DataFrame(df_cases)
df_cases = df_cases.reset_index() 

df_harmed = pd.DataFrame(df_harmed)
df_harmed = df_harmed.reset_index() 

df_kills = pd.DataFrame(df_kills)
df_kills = df_kills.reset_index() 


df_trend = pd.concat([df_kills,df_harmed['n_injured'],df_cases['incident_id']],axis=1)
df_trend = df_trend.rename(columns={"n_killed": "Total Number of People killed", "n_injured": "Total Number of People injured",'incident_id':'Total Number of Incidents'})
```

Picture on a weekly basis

```python
df_trend2 = pd.melt(df_trend, id_vars = ['datetime'] )
df_weekly = df_trend2.set_index('datetime')
df_weekly = df_weekly.groupby('variable').resample('W').sum()
df_weekly = df_weekly.reset_index()
df_weekly = df_weekly[df_weekly['value']!= 0]

alt.Chart(df_weekly[df_weekly['datetime'] >= '01-01-2014']).mark_line().encode(
    x=alt.X('datetime:T', title = 'Time', axis = alt.Axis(format='%Y' , tickCount= 5),
 ),
    y= alt.Y('value:Q', title = 'Absolute Number of Cases per Week'),
    color= alt.Color('variable', title = 'Variable of Interest', )
).properties(width=700, height = 400,title = 'Gun Violence in the US - Weekly Rates').configure_view(
    stroke='transparent'
).configure_legend(orient = 'bottom')
```

Picture on a daily basis

```python
df_trend2 = pd.melt(df_trend, id_vars = ['datetime'] )
df_weekly = df_trend2.set_index('datetime')
df_weekly = df_weekly.groupby('variable').resample('D').sum()
df_weekly = df_weekly.reset_index()
df_weekly = df_weekly[df_weekly['value']!= 0]

alt.Chart(df_weekly[df_weekly['datetime'] >= '01-01-2014']).mark_line().encode(
    x=alt.X('datetime:T', title = 'Time', axis = alt.Axis(format='%Y' , tickCount= 5),
 ),
    y= alt.Y('value:Q', title = 'Absolute Number of Cases per Day'),
    color= alt.Color('variable', title = 'Variable of Interest', )
).properties(width=700, height = 400, title = 'Gun Violence in the US - Daily Rates').configure_view(
    stroke='transparent'
).configure_legend(orient = 'bottom')
```

This looks better. 

However, there are a lot of datapoints (since there are no days where the total amount of incidents is below 50). Maybe I should narrow down my plot and inspect aggregated levels on year, month and day basis

```python
df['weekday'] = df['datetime'].dt.day_name()
df['month'] = df['datetime'].dt.month_name()
df['year'] = df['datetime'].dt.year
df['weekday'] = df['weekday'].astype('object')
df['month'] = df['month'].astype('object')
df['year'] = df['year'].astype('object')
```

A. Changes over Years

```python
df_years = df.groupby('year').agg('sum')[['n_killed','n_injured']]
df_years = df_years.rename(index = str, columns={"n_killed":"People Killed", "n_injured":"People Injured"})
df_years = df_years.reset_index()  
df_years = pd.melt(df_years, id_vars = ['year'], value_vars = ['People Killed', 'People Injured'])

df_months = df.groupby('month').agg('sum')[['n_killed','n_injured']]
df_months = df_months.rename(index = str, columns={"n_killed":"People Killed", "n_injured":"People Injured"})
df_months = df_months.reset_index()  
df_months = pd.melt(df_months, id_vars = ['month'], value_vars = ['People Killed', 'People Injured'])

df_days = df.groupby('weekday').agg('sum')[['n_killed','n_injured']]
df_days = df_days.rename(index = str, columns={"n_killed":"People Killed", "n_injured":"People Injured"})
df_days = df_days.reset_index()  
df_days = pd.melt(df_days, id_vars = ['weekday'], value_vars = ['People Killed', 'People Injured'])
```

### 1.2.1 Identifying Holidays

```python
df_freq = pd.crosstab(df['date'], columns='count')
df_freq['count'].sort_values(ascending = False)
```

There is a heavy spike on 01.01.2017 with a very high number of 342 cases

To extract regular occuring (holiday) events, I will take the subset of the 200 days with the highest amount of cases and look which repeat every year

```python
df_freq = df_freq.reset_index() 
df_freq['datetime'] = pd.to_datetime(df_freq['date'])# Making sure that there is a datetime format
df_freq['datetstr'] = df_freq['datetime'].apply(lambda x: x.strftime('%Y-%m-%d')) #Generating string from datetime-format
df_freq['Month-day'] = df_freq['datetstr'].str[5:] #Removing the year

df_freq_200 = df_freq.sort_values(by = 'count', ascending = False)[0:200] ##Only extracting the 200 days with the higesht number of cases
pd.crosstab(df_freq_200['Month-day'], columns='count').sort_values(by = 'count',ascending = False)
```

Ok, here I can see that there are several holidays with an increase in cases:
- January 01 (New Year)
- July 04 (Indepence Day) (and 05.07) 
- September 05 (Maybe Labor Day?)

## Controlling Using Number of Deaths

```python
df_18 =df[df['year']!=2013]
df_ts = df_18[['n_killed','date']]
df_ts.index=df_18['date']

df_date_index = df_ts[['date','n_killed']]
df_date_index = df_date_index.set_index('date')
df_prophet_deaths = df_date_index.copy()
df_prophet_deaths.reset_index(drop=False,inplace=True)
df_prophet_deaths.columns = ['ds','y']
df_prophet_deaths = df_prophet_deaths[df_prophet_deaths['y'] <= 10]

my_model_deaths = Prophet(interval_width=0.95, daily_seasonality= False)
my_model_deaths.add_country_holidays(country_name='US')
my_model_deaths.fit(df_prophet_deaths)
future_deaths = my_model_deaths.make_future_dataframe(periods=720,freq='D')
forecast_deaths = my_model_deaths.predict(future_deaths)
fig = my_model_deaths.plot(forecast_deaths)
```
